```python
#for each lowercase word, I want to record what word comes after it
#and with what frequency

#start by creating a standard dictionary
big_dict = defaultdict(list)

for answer in cards:
    word_bank = answer.split()
    
    #get every word that comes after it in a dictionary
    if len(word_bank) > 1:
        for idx, word in enumerate(word_bank[:-1]):
            big_dict[word].append(word_bank[idx + 1])
    else:
        big_dict[word].append(word_bank[idx + 1])
```

```python
big_dict
```

```python
random.choice(list(big_dict.keys()))
```

```python
def make_card():
    
    #get a random starting word
    first_word = random.choice(list(big_dict.keys()))
    sentence = [first_word]
    
    while '.' not in sentence[-1]:
        next_word = random.choice(big_dict[sentence[-1]])
        sentence.append(next_word)
        
    return ' '.join(sentence) 
```
