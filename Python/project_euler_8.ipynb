Actually, instead of iterating over every single number and remultiplying all 13 values, I can divide by the number right before it the len 13 window, and just multiply by the new number as the window moves

```python
max_prod = 0
previous = 1

for i in range(4):
    previous *= int(num_str[i])


for i in range(1,len(num_str) - 4):
    
    #if 0 is in the len 13 span, everything will be 0, so skip those products
    if '0' in num_str[i:i+4] or ('0' == num_str[i]) or ('0' == num_str[i+4]):
        continue

    else:
        print(i)
        #get the possible new max
        previous = int((previous/int(num_str[i]))*int(num_str[i+4]))
        print('previous')
        print(previous)
        print()

        if previous > max_prod:
            max_prod = previous
            print('new_max')
            print(max_prod)
            print()
        else:
            continue
            
print('Max Product =', max_prod)
```
