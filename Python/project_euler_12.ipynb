The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1  
 3: 1,3  
 6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28  
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
```

```python
#well I know that the sum of the first n natural numbers is n(n+1)/2 so maybe I can do something with that

def sum_up_to_n(n):
    return int(n*(n+1)/2)


#write a function to count the divisors
def count_factors(n):
    total = 0
    
    for i in range(1,math.ceil(n**(1/2))+1):
        #this range miht give an issue for the smaller numbers, but it shortens the runtime like CRAZY, 
        # and I'm only giving it n>7 so I think it'll be fine
        if n % i == 0:
            total += 2
    return total
```

```python
less_than_500 = True
n = 7

while less_than_500:
    if count_factors(sum_up_to_n(n)) >= 500:
        print('The first triangle number with more than 500 divisors is', n)
        print('The sum up to n =', sum_up_to_n(n))
        print('The number of divisors =', count_factors(sum_up_to_n(n)))
        
        break
        
    
    n += 1
    
```
